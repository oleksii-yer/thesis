```python
import cupy as cp
import numpy as np
import time

# Create large arrays with NumPy and CuPy
size = 10**8  # Array size for testing

# Create arrays
np_array = np.random.rand(size)
cp_array = cp.random.rand(size)

# Measure time for NumPy operation
start_time = time.time()
np_sum = np.sum(np_array)
np_time = time.time() - start_time
print(f"NumPy sum took: {np_time:.6f} seconds")

# Measure time for CuPy operation (on GPU)
start_time = time.time()
cp_sum = cp.sum(cp_array)
cp_time = time.time() - start_time
print(f"CuPy sum took: {cp_time:.6f} seconds")

# Transfer the result back to CPU if needed
cp_sum_cpu = cp_sum.get()

# Compare results
print(f"NumPy sum: {np_sum}")
print(f"CuPy sum: {cp_sum_cpu}")
```
